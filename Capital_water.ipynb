```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams.update({'font.size': 12})
plt.style.use('ggplot')
```

```python
Csv_Path = "/Users/dongmintian994410/Jupyter/Capital_water/无标题电子表格 - TRANMH19 (1).csv"
Capital_water = pd.read_csv(Csv_Path, engine='python', skipfooter=3)
Capital_water['Date'] = pd.to_datetime(Capital_water['Date'], format='%Y-%m-%d %H:%M')
Capital_water.set_index(['Date'], inplace=True)
```

```python
Capital_water.plot()
plt.ylabel('Monthly airline passengers (x1000)')
plt.xlabel('Date')
plt.show()
```
